```python
import torch
import matplotlib.pyplot as plt
```

```python
with open("day_head_circumference.csv") as f:
    data = [i.split(",") for i in f.readlines()]

days = [float(r[0]) for r in data[1:]]
head_circumference = [float(r[1]) for r in data[1:]]

x_train = torch.tensor(days).reshape(-1, 1)
y_train = torch.tensor(head_circumference).reshape(-1, 1)
```

```python
# plot
plt.figure(figsize=(10, 6))
plt.scatter(days, head_circumference, c='blue', marker='o', label='Data Points')
plt.xlabel('days')
plt.ylabel('head_circumference')
plt.title('Day and head circumference')
plt.legend()
plt.grid(True)
plt.show()
```

```python
class NonLinearRegressionModel:
    def __init__(self):
        self.W = torch.tensor([[0.0]], requires_grad=True)  
        self.b = torch.tensor([[0.0]], requires_grad=True)

    def f(self, x):
        return 20 *  torch.sigmoid((x @ self.W + self.b)) + 31 
    
    # Uses Mean Squared Error
    def loss(self, x, y):
        return torch.mean(torch.square(self.f(x) - y))
```

```python
model = NonLinearRegressionModel()
optimizer = torch.optim.SGD([model.W, model.b], lr=0.000001)

epochs = 10_000
for epoch in range(epochs):
    if epoch % (epochs // 10)  == 0:
        print("Done:", epoch)
    model.loss(x_train, y_train).backward()
    optimizer.step()
    optimizer.zero_grad()

print("DONE !!!")
```

```python
print("W = %s, b = %s, loss = %s" % (model.W, model.b, model.loss(x_train, y_train)))

plt.plot(x_train, y_train, 'o', label='$(x^{(i)},y^{(i)})$')
plt.xlabel('days old')
plt.ylabel('head circumference')


x_entries = torch.linspace(torch.min(x_train), torch.max(x_train), steps=750).reshape(-1, 1)
y_entries = model.f(x_entries).detach()

plt.plot(x_entries, y_entries, label='$f(x) = 20\\sigma(xW + b) + 31$')
plt.legend()
plt.show()
```

```python
import numpy
from matplotlib import pyplot as plt
```

```python
def f(Re, Im, m):
    c = complex(Re, Im)
    z = complex(0, 0)
    for i in range(m):
        z = z*z + c
        if z.real ** 2 + z.imag ** 2 >= 2:
            return i
    return m
```

```python
COLS = ROWS = 2000

res = numpy.zeros([ROWS, COLS])
# Real axis in the argand plane
for r_ind, Re in enumerate(numpy.linspace(-2, 1, ROWS)):
    # Imaginary axis in the argand plane
    for i_ind, Im in enumerate(numpy.linspace(-2, 1, COLS)):
        res[r_ind, i_ind] = f(Re, Im, 500)
```

```python
plt.figure(dpi=100)
plt.imshow(res.T, cmap="hot", interpolation="bilinear", extent=[-2, 1, -1, 1])
plt.show()
```
